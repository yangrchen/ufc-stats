```python
import requests
import pandas as pd
import yaml
from bs4 import BeautifulSoup
from pathlib import Path
from operator import methodcaller

with open('config.yaml', 'r') as f:
    config = yaml.load(f, Loader=yaml.FullLoader)
```

```python
req = requests.get(config['completed_events_all_url'])

if req.ok:
    soup = BeautifulSoup(req.content, 'html.parser') 
```

```python
stats_table = soup.find('tbody')
rows = stats_table.find_all('tr', {'class': 'b-statistics__table-row'})

event_links = [row.find('a').get('href') for row in rows[1:]]

event_links
```

```python
event_path = Path('event')
if not event_path.exists():
    event_path.mkdir()
    for idx, event in enumerate(event_links):
        req = requests.get(event)
        with open(event_path / f'event_{idx}', 'wb') as f:
            f.write(req.content)
```

```python
def retrieve_fight_links(path: Path) -> list:
    all_fight_links = []
    for event in path.glob('*'):
        with open(event, 'rb') as f:
            content = f.read()
            soup = BeautifulSoup(content, 'html.parser')
            fight_links = soup.find_all('a', {'class': 'b-flag_style_green'})
            fight_links = list(map(methodcaller('get', 'href'), fight_links))
            all_fight_links.extend(fight_links)
    return all_fight_links
```
